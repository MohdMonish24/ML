# Q8. What is R -squared and adjusted R -square?

#  R -Coefficient of determination

The R score, also known as the coefficient of determination or goodness of fit, is a metric in machine learning that measures how well a regression model fits the data

A higher R score indicates a better fit, and it allows us to compare the models performance against a baseline model that uses the mean value of the target variable.

If the R score is zero, it means that the regression line is equivalent to the mean line. In this case, both lines overlap, indicating that the models performance is poor and it fails to capture any meaningful relationship in the data.

If the R score is one, it implies that the regression line perfectly fits the data without any errors. However, achieving a perfect R score is highly unlikely in real-world scenarios.

In general cases, the R score falls between zero and one, such as 0.8. This indicates that the model can explain 80 percent of the variance in the data. As the R score approaches one, the regression line gets closer to perfection, indicating an improvement in model performance.

Few Important points Related to R

- Ris a measure of how well a regression model fits the data and explains the variability in the dependent variable.

- It ranges between 0 and 1, with 0 indicating no variability explained and 1 indicating a perfect fit.

- R can be used to compare different models, with higher values indicating better performance.

- However, Rshould be interpreted with caution as it doesnt provide information on statistical significance or causality.







#  Adjusted R2  or Adjusted R-Squared

Adjusted R-squared is a modified version of R-squared that has been adjusted for the 
number of predictors in the model. The adjusted R-squared increases when the new term 
improves the model more than would be expected by chance. It decreases when a 
predictor improves the model by less than expected. Typically, the adjusted R-squared is 
positive, not negative. It is always lower than the R-squared



Few Important points Related to R

- Adjusted R is a modified version of the R (R-squared) score that adjusts for the number of features in a regression model.

- Unlike R, which tends to increase as more features are added, Adjusted R penalizes the inclusion of unnecessary features that do not contribute significantly to the models predictive power.

- Adjusted R provides a more reliable measure of model performance by accounting for the trade-off between model complexity and goodness of fit.

- It is particularly useful when comparing regression models with different numbers of features, as it helps identify the model that strikes the right balance between simplicity and predictive power.

- The Adjusted R score ranges from negative infinity to 1, with a higher score indicating a better fit and a greater proportion of variance explained by the independent variables.

- Negative values of Adjusted R may occur when the model performs worse than a simple average or when the number of features is excessively large compared to the sample size.

# Q1. What is regularization and why is it used in machine learning?

Regularization is a technique used in machine learning to prevent overfitting and improve the generalization ability of models. It involves adding a penalty term to the loss function during the training process, which helps control the complexity of the model. By adding this penalty, regularization discourages the model from fitting noise and random fluctuations in the training data.

The goal of regularization is to find the right balance between fitting the training data well and avoiding overfitting. It encourages the model to capture the underlying patterns in the data while avoiding excessive complexity.

Regularization Techniques:

1- L1 regularization (Lasso)

2- L2 regularization (Ridge).

3- Elatic Net



Ridge, Lasso, and Elastic Net are three popular regularization techniques used in machine learning to address overfitting and improve the performance of models.


#        :-

1   : Regularization prevents overfitting, enhances generalization, and promotes better feature selection. It's an invaluable tool to achieve accurate and robust models.

2   : L1 (Lasso), L2 (Ridge), and Elastic Net regularization each bring unique strengths. L1 encourages feature selection, L2 discourages extreme parameter values, while Elastic Net strikes a balance between both.

3    : Regularization is ideal when the training data is limited, there are many features, or overfitting is observed. It's widely applicable in linear regression, logistic regression, neural networks, and support vector machines.

# 1- Ridge Regularization:

Ridge regularization, also known as L2 regularization, adds the sum of squared coefficients to the loss function as penalty.



This penalty term encourages the model to have smaller and more balanced weights.

Ridge regularization is effective when dealing with multicollinearity(statistical concept where several independent variables in a model are correlated), where features are highly correlated.

# 2- Lasso Regularization:

Lasso regularization, also known as L1 regularization, adds the sum of absolute values of coefficients to the loss function.



Lasso encourages sparsity by bringing some coefficients to exactly zero.

It also performs feature selection, effectively selecting the most important features and disregarding irrelevant ones. Lasso is particularly useful when dealing with high-dimensional datasets.

# 3- Elastic Net Regularization (L1+L2):

Elastic Net regularization combines the properties of both Ridge and Lasso regularizationimage.pngIt adds a combination of the L1 and L2 penalties to the loss function.

Elastic Net is beneficial when dealing with datasets that have a large number of features and multicollinearity issues. It provides a balance between feature selection (Lasso) and handling correlated features (Ridge). Conclusion :

# Q1. What is Logistic Regression?

Logistic regression is a statistical method used for predicting binary outcomes (i.e. outcomes
with two possible values, such as success or failure). It is a popular technique for many
applications, such as predicting customer churn or whether a patient has a certain disease.

 Its a Supervised Learning technique

 It is used for predicting the categorical dependent variable using a given set of independent variables.

 Logistic regression predicts the output of a categorical dependent variable.

 The outcome must be a categorical or discrete value.

 It can be either Yes or No, 0 or 1, true or False, etc.

 It gives the probabilistic values which lie between 0 and 1.

 Logistic Regression is much like Linear Regression except that how they are used.

 Linear Regression solves Regression problems, whereas Logistic regression solves the
classification problems.

 In Logistic Regression, we don't directly fit a straight line to our data like in linear
regression.

 In Logistic Regression, we fit an S shaped curve, called Sigmoid, to our data/observations.

 Logistic Regression is a significant machine learning algorithm.

 It can provide probabilities and classify new data using continuous and discrete datasets.

 Logistic Regression can be used to classify the observations using different types of data
and can easily determine the most effective variables used for the classification.

# Type of Logistic Regression:

Based on the categories, Logistic Regression can be classified into three types:

1. Binomial: In binomial Logistic regression, there can be only two possible types of the 
dependent variables, such as 0 or 1, Pass or Fail, etc.

2. Multinomial: In multinomial Logistic regression, there can be 3 or more possible unordered 
types of the dependent variable, such as "cat", "dogs", or "sheep"

3. Ordinal: In ordinal Logistic regression, there can be 3 or more possible ordered types of 
dependent variables, such as "low", "Medium", or "High"

# How Does Logistic Regression Work?

Logistic regression works by building a model that predicts the probability of the outcome 
being a certain value (e.g. success or failure). This probability is then used to make a 
prediction about the outcome. The model is built by finding the best parameters (also known 
as coefficients) that minimize the error between the predicted probabilities and the actual 
outcomes.
